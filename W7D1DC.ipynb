```python
# For the tasks below, do them using a regular Python list implementation, and then use NumPy.

#     1. Create a table (i.e. a 2d array) of size M x N filled with random integers between 1 and 100,
#     where 1 < N < 40 and 1 < M < 50.

# Python:
from random import randint


N = randint(1,40)
M = randint(1,50)

table = [[randint(1,100) for j in range(N)] for i in range(M)]


```

```python
# numpy:
from numpy import random

x = random.randint(1,100, size=(random.randint(1,50),random.randint(1,40)))
```

```python
print(x)
```

```python
# 2. Print out the third row

# Python:
print(table[2])
```

```python
# numpy:
print(x[2,:])
```

```python
# 3. Print out the third column 

# Python:
print([i[2] for i in table])
```

```python
# numpy:
print(x[:,2])
```

```python
# 4. Set every element in the last row equal to 7

# Python:
table[-1] = [7 for j in range(N)]

print(table[-1])
```

```python
# numpy:
x[-1] = 7
print(x)
```

```python
# 5. Set every element in the last column equal the sum of the first two columns. (note: the result of
# the sum is a list which will the same length as the last column)

# Python:
for i, ival in enumerate(table):
    table[i][-1] = ival[0] + ival[1]
print([i[-1] for i in table])
```

```python
# numpy:
x[:,-1]=x[:,0]+x[:,1]
print(x[:,-1])
```
